```python
import pandas as pd
```

```python
df_raw = pd.read_csv('raw.csv')
df_norm = pd.read_csv('with_norm.csv')
```

```python
df_raw_nlp = (
    df_raw
    .rename(columns={'Customer Review': 'text'})
    [['Category', 'Product Name', 'Customer Rating', 'Sentiment', 'Emotion', 'text']]
    .copy()
)
```

```python
df_norm_nlp = (
    df_norm
    .rename(columns={'text_norm': 'text'})
    [['Category', 'Product Name', 'Customer Rating', 'Sentiment', 'Emotion', 'text']]
    .copy()
)
```

```python
df_raw_nlp.head(3)
```

```python
df_norm_nlp.head(3)
```

```python
from sklearn.model_selection import train_test_split

RANDOM_STATE = 42
```

raw

```python
train_df_raw, temp_df_raw = train_test_split(
    df_raw_nlp,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=df_raw_nlp['Sentiment']
)

val_df_raw, test_df_raw = train_test_split(
    temp_df_raw,
    test_size=0.5,
    random_state=42,
    shuffle=True,
    stratify=temp_df_raw['Sentiment']
)
```

```python
print(len(train_df_raw), len(val_df_raw), len(test_df_raw))
```

```python
train_df_raw.to_csv("train_raw.csv", index=False)
val_df_raw.to_csv("val_raw.csv", index=False)
test_df_raw.to_csv("test_raw.csv", index=False)
```

norm

```python
train_df_norm, temp_df_norm = train_test_split(
    df_norm_nlp,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=df_norm_nlp['Sentiment']
)

val_df_norm, test_df_norm = train_test_split(
    temp_df_norm,
    test_size=0.5,
    random_state=42,
    shuffle=True,
    stratify=temp_df_norm['Sentiment']
)
```

```python
print(len(train_df_norm), len(val_df_norm), len(test_df_norm))
```

```python
train_df_norm.to_csv("train_norm.csv", index=False)
val_df_norm.to_csv("val_norm.csv", index=False)
test_df_norm.to_csv("test_norm.csv", index=False)
```
